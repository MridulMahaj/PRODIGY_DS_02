```python
# Install necessary libraries (if not already installed)
!pip install pandas numpy matplotlib seaborn

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset (assuming it's available in the same directory or provide the path)
# Replace 'path_to_dataset' with the actual path or URL if downloading from Kaggle
df = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')

# Display the first few rows and basic info about the dataset
df.head()
```

```python
# Checking for missing values
print("Missing values:")
print(df.isnull().sum())

# Handling missing values (example: filling missing values in 'Age' with median)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Dropping unnecessary columns if needed
# df.drop(['Cabin', 'Ticket'], axis=1, inplace=True)

# Checking data types and converting if necessary (e.g., converting categorical to numeric)
# df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Handling outliers if necessary (e.g., using quantiles)
# q_low = df['column_name'].quantile(0.01)
# q_high  = df['column_name'].quantile(0.99)
# df_filtered = df[(df['column_name'] < q_high) & (df['column_name'] > q_low)]

# Confirming data integrity
print("\nCleaned dataset:")
print(df.head())
```

```python
# Summary statistics
print("\nSummary statistics:")
print(df.describe())

# Select only numerical columns for correlation
numerical_df = df.select_dtypes(include=['number'])

# Visualize distributions and relationships (examples)
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare distribution by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()

# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', linewidths=.5) # Calculate correlation on numerical data only
plt.title('Correlation Matrix')
plt.show()
```
